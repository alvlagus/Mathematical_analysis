## –ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–æ–µ –∑–∞–¥–∞–Ω–∏–µ –∫ —É—Ä–æ–∫—É 7: "–ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏–∏ –æ–¥–Ω–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π. –ß–∞—Å—Ç—å 2"

```python
# –ø–æ–¥–≥—Ä—É–∂–∞–µ–º—ã–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import splev, splrep
import math
```

```python
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.splrep.html


x = np.linspace(0, 10, 10)
y = np.sin(x)
plt.plot(x, y, marker='o', ls='')
spl = splrep(x,y)
x2 = np.linspace(0, 10, 200)
y2 = splev(x2, spl)

plt.plot(x, y, 'o', x2, y2)
plt.show()
```

**1. –ù–∞–π—Ç–∏ –¥–ª–∏–Ω—É ùë• –∏ —à–∏—Ä–∏–Ω—É ùë¶ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–∞ –ø—Ä–∏ –∑–∞–¥–∞–Ω–Ω–æ–º –ø–µ—Ä–∏–º–µ—Ç—Ä–µ ùëÉ=144 c–º, –ø—Ä–∏ –∫–æ—Ç–æ—Ä—ã—Ö –¥–∞–Ω–Ω—ã–π –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ –∏–º–µ–µ—Ç –Ω–∞–∏–±–æ–ª—å—à—É—é –ø–ª–æ—â–∞–¥—å ùëÜ**



x - –¥–ª–∏–Ω–∞ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–∞, —Ç–æ–≥–¥–∞ y = 72-x - –µ–≥–æ —à–∏—Ä–∏–Ω–∞.  
–ü–æ —É—Å–ª–æ–≤–∏—é –∑–∞–¥–∞—á–∏, $P = 2x + 2y = 144$ —Å–º.  
$S = x \cdot y$  
$S = x \cdot (72 - x) = 72x - x^2$  
$S'(x) = (72x - x^2)' = 72 - 2x$  
–ù–∞–π–¥–µ–º —ç–∫—Å—Ç—Ä–µ–º—É–º —Ñ—É–Ω–∫—Ü–∏–∏ S(x).  
–ù–µ–æ–±—Ö–æ–¥–∏–º–æ–µ —É—Å–ª–æ–≤–∏–µ:  
$S'(x) = 72 - 2x = 0$  
$72 - 2x = 0$  
$2x = 72$  
$x = 36$  
–ù–µ–æ–±—Ö–æ–¥–º–æ–µ —É—Å–ª–æ–≤–∏–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è.  

–î–æ—Å—Ç–∞—Ç–æ—á–Ω–æ–µ —É—Å–ª–æ–≤–∏–µ:  
$S'(35) = 72 - 2x = 72 - 2 \cdot 35 = 2 > 0$  
$S'(37) = 72 - 2x = 72 - 2 \cdot 37 = -2 < 0$  
  
–°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, –ø—Ä–∏ –ø–µ—Ä–µ—Ö–æ–¥–µ —á–µ—Ä–µ–∑ —Ç–æ—á–∫—É $x_0 = 36$, S'(x) –º–µ–Ω—è–µ—Ç –∑–Ω–∞–∫. –î–æ—Å—Ç–∞—Ç–æ—á–Ω–æ–µ —É—Å–ª–æ–≤–∏–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è. –°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ —Ç–æ—á–∫–∞ $x_0$ —è–≤–ª—è–µ—Ç—Å—è —Ç–æ—á–∫–æ–π —ç–∫—Å—Ç—Ä–µ–º—É–º–∞. –¢–æ—á–∫–æ–π –º–∞–∫—Å–∏–º—É–º–∞.  
–¢–æ–≥–¥–∞,  
$y = 72-x = 72-36 = 36$  
$S = x \cdot y = 36 \cdot 36 = 1296$ —Å–º


**2. –ù–∞–π—Ç–∏ —ç–∫—Å—Ç—Ä–µ–º—É–º—ã —Ñ—É–Ω–∫—Ü–∏–π (–µ—Å–ª–∏ –æ–Ω–∏ –µ—Å—Ç—å)**

a. $$y = |2x|$$

$y = |2x|$  
$y' = (|2x|)' = \frac{2x}{|x|}$  
*–ù–µ–æ–±—Ö–æ–¥–∏–º–æ–µ —É—Å–ª–æ–≤–∏–µ —ç–∫—Å—Ç—Ä–µ–º—É–º–∞.*  
–ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –æ—Ç |—Ö| –ø—Ä–∏ —Ö<0 = -1. –ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –æ—Ç |—Ö| –ø—Ä–∏ —Ö>0 = 1. –ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –æ—Ç |—Ö| –ø—Ä–∏ —Ö=0 –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç, —Ç.–∫. –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ –ø—Ä–æ–≤–µ—Å—Ç–∏ –∫–∞—Å–∞—Ç–µ–ª—å–Ω—É—é –≤ —ç—Ç–æ–π —Ç–æ—á–∫–µ –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω—ã–º –æ–±—Ä–∞–∑–æ–º.  
–°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, –Ω–µ–æ–±—Ö–æ–¥–º–æ–µ —É—Å–ª–æ–≤–∏–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è.

*–î–æ—Å—Ç–∞—Ç–æ—á–Ω–æ–µ —É—Å–ª–æ–≤–∏–µ —ç–∫—Å—Ç—Ä–µ–º—É–º–∞.*  
$y'(-1) = \frac{2(-1)}{|-1|} = -2 < 0$  
$y'(1) = \frac{2(1)}{|1|} = 2 > 0$  
  
–°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, –ø—Ä–∏ –ø–µ—Ä–µ—Ö–æ–¥–µ —á–µ—Ä–µ–∑ —Ç–æ—á–∫—É $x_0 = 0$, y'(x) –º–µ–Ω—è–µ—Ç –∑–Ω–∞–∫. –î–æ—Å—Ç–∞—Ç–æ—á–Ω–æ–µ —É—Å–ª–æ–≤–∏–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è. –°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ —Ç–æ—á–∫–∞ $x_0$ —è–≤–ª—è–µ—Ç—Å—è —Ç–æ—á–∫–æ–π —ç–∫—Å—Ç—Ä–µ–º—É–º–∞. –¢–æ—á–∫–æ–π –º–∏–Ω–∏–º—É–º–∞.  

```python
def f(x):
    return abs(2*x)

x = np.linspace(-1, 1, 100)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

b. $$y = x^3$$

$y = x^3$  
$y' = (x^3)' = 3x^2$  
*–ù–µ–æ–±—Ö–æ–¥–∏–º–æ–µ —É—Å–ª–æ–≤–∏–µ —ç–∫—Å—Ç—Ä–µ–º—É–º–∞.*  
$3x^2 = 0$  
$x = 0$  
–ù–µ–æ–±—Ö–æ–¥–º–æ–µ —É—Å–ª–æ–≤–∏–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è.

*–î–æ—Å—Ç–∞—Ç–æ—á–Ω–æ–µ —É—Å–ª–æ–≤–∏–µ —ç–∫—Å—Ç—Ä–µ–º—É–º–∞.*  
$y'(-1) = 3 \cdot (-1)^2 = 3 > 0$  
$y'(1) = 3 \cdot (1)^2 = 3 > 0$  
  
–°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, –ø—Ä–∏ –ø–µ—Ä–µ—Ö–æ–¥–µ —á–µ—Ä–µ–∑ —Ç–æ—á–∫—É $x_0 = 0$, y'(x) –Ω–µ –º–µ–Ω—è–µ—Ç –∑–Ω–∞–∫. –î–æ—Å—Ç–∞—Ç–æ—á–Ω–æ–µ —É—Å–ª–æ–≤–∏–µ –Ω–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è. –°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ —Ç–æ—á–∫–∞ $x_0$ –Ω–µ —è–≤–ª—è–µ—Ç—Å—è —Ç–æ—á–∫–æ–π —ç–∫—Å—Ç—Ä–µ–º—É–º–∞.

```python
def f(x):
    return pow(x,3)

x = np.linspace(-5, 5, 100)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

c. $$y = e^{3x}$$

$y = e^{3x}$  
$y' = (e^{3x})' = 3e^{3x}$  

*–ù–µ–æ–±—Ö–æ–¥–∏–º–æ–µ —É—Å–ª–æ–≤–∏–µ —ç–∫—Å—Ç—Ä–µ–º—É–º–∞.*  
$3e^{3x} \ne 0$  
$y' = 3e^{3x}$ –±–æ–ª—å—à–µ –Ω—É–ª—è –ø—Ä–∏ –ª—é–±–æ–º —Ö –∏ –Ω–µ—Ç —Ç–æ—á–µ–∫, –≤ –∫–æ—Ç–æ—Ä—ã—Ö –±—ã –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –æ—Ç—Å—É—Ç—Å—Ç–≤–æ–≤–∞–ª–∞.  
–°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ–µ —É—Å–ª–æ–≤–∏–µ —ç–∫—Å—Ç—Ä–µ–º—É–º–∞ –Ω–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è. –≠–∫—Å—Ç—Ä–µ–º—É–º –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç. 

```python
def f(x):
    return pow(math.e,(3*x))

x = np.linspace(-10, 10, 1000)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

d. $$y = x^3 - 5x$$

$y = x^3 - 5x$
$y' = (x^3 - 5x)' = 3x^2 - 5$  

*–ù–µ–æ–±—Ö–æ–¥–∏–º–æ–µ —É—Å–ª–æ–≤–∏–µ —ç–∫—Å—Ç—Ä–µ–º—É–º–∞.*  
$3x^2 - 5 = 0$  
$x^2 = \frac{5}{3}$  
$x_1 = -\sqrt{\frac{5}{3}} \approx -1,3$  
$x_2 = \sqrt{\frac{5}{3}} \approx 1,3$  
–ù–µ–æ–±—Ö–æ–¥–º–æ–µ —É—Å–ª–æ–≤–∏–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è.  

*–î–æ—Å—Ç–∞—Ç–æ—á–Ω–æ–µ —É—Å–ª–æ–≤–∏–µ —ç–∫—Å—Ç—Ä–µ–º—É–º–∞.*  
$y'(-2) = 3 \cdot (-2)^2 - 5 = 7 > 0$  
$y'(0) = 3 \cdot (0)^2 - 5 = -5 < 0$  
$y'(2) = 3 \cdot (2)^2 - 5 = 7 > 0$  

–°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, –ø—Ä–∏ –ø–µ—Ä–µ—Ö–æ–¥–µ —á–µ—Ä–µ–∑ —Ç–æ—á–∫–∏ $x_1 = -\sqrt{\frac{5}{3}}$ –∏ $x_2 = \sqrt{\frac{5}{3}}$, y'(x) –º–µ–Ω—è–µ—Ç –∑–Ω–∞–∫. –î–æ—Å—Ç–∞—Ç–æ—á–Ω–æ–µ —É—Å–ª–æ–≤–∏–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è. –°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ —Ç–æ—á–∫–∏ $x_1, x_2$ —è–≤–ª—è—é—Ç—Å—è —Ç–æ—á–∫–∞–º–∏ —ç–∫—Å—Ç—Ä–µ–º—É–º–∞. –¢–æ—á–∫–æ–π –º–∞–∫—Å–∏–º—É–º–∞ –∏ –º–∏–Ω–∏–º—É–º–∞, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ.  

```python
def f(x):
    return (pow(x, 3))-5*x

x = np.linspace(-3, 3, 100)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```
