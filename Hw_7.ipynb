**1. Найти длину 𝑥 и ширину 𝑦 прямоугольника при заданном периметре 𝑃=144 cм, при которых данный прямоугольник имеет наибольшую площадь 𝑆**



x - длина прямоугольника, тогда y = 72-x - его ширина.  
По условию задачи, $P = 2x + 2y = 144$ см.  
$S = x \cdot y$  
$S = x \cdot (72 - x) = 72x - x^2$  
$S'(x) = (72x - x^2)' = 72 - 2x$  
Найдем экстремум функции S(x).  
Необходимое условие:  
$S'(x) = 72 - 2x = 0$  
$72 - 2x = 0$  
$2x = 72$  
$x = 36$  
Необходмое условие выполняется.  

Достаточное условие:  
$S'(35) = 72 - 2x = 72 - 2 \cdot 35 = 2 > 0$  
$S'(37) = 72 - 2x = 72 - 2 \cdot 37 = -2 < 0$  
  
Следовательно, при переходе через точку $x_0 = 36$, S'(x) меняет знак. Достаточное условие выполняется. Следовательно точка $x_0$ является точкой экстремума. Точкой максимума.  
Тогда,  
$y = 72-x = 72-36 = 36$  
$S = x \cdot y = 36 \cdot 36 = 1296$ см


**2. Найти экстремумы функций (если они есть)**

a. $$y = |2x|$$

$y = |2x|$  
$y' = (|2x|)' = \frac{2x}{|x|}$  
*Необходимое условие экстремума.*  
Производная от |х| при х<0 = -1. Производная от |х| при х>0 = 1. Производная от |х| при х=0 не существует, т.к. невозможно провести касательную в этой точке единственным образом.  
Следовательно, необходмое условие выполняется.

*Достаточное условие экстремума.*  
$y'(-1) = \frac{2(-1)}{|-1|} = -2 < 0$  
$y'(1) = \frac{2(1)}{|1|} = 2 > 0$  
  
Следовательно, при переходе через точку $x_0 = 0$, y'(x) меняет знак. Достаточное условие выполняется. Следовательно точка $x_0$ является точкой экстремума. Точкой минимума.  

```python
def f(x):
    return abs(2*x)

x = np.linspace(-1, 1, 100)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

b. $$y = x^3$$

$y = x^3$  
$y' = (x^3)' = 3x^2$  
*Необходимое условие экстремума.*  
$3x^2 = 0$  
$x = 0$  
Необходмое условие выполняется.

*Достаточное условие экстремума.*  
$y'(-1) = 3 \cdot (-1)^2 = 3 > 0$  
$y'(1) = 3 \cdot (1)^2 = 3 > 0$  
  
Следовательно, при переходе через точку $x_0 = 0$, y'(x) не меняет знак. Достаточное условие не выполняется. Следовательно точка $x_0$ не является точкой экстремума.

```python
def f(x):
    return pow(x,3)

x = np.linspace(-5, 5, 100)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

c. $$y = e^{3x}$$

$y = e^{3x}$  
$y' = (e^{3x})' = 3e^{3x}$  

*Необходимое условие экстремума.*  
$3e^{3x} \ne 0$  
$y' = 3e^{3x}$ больше нуля при любом х и нет точек, в которых бы производная отсутствовала.  
Следовательно, необходимое условие экстремума не выполняется. Экстремум отсутствует. 

```python
def f(x):
    return pow(math.e,(3*x))

x = np.linspace(-10, 10, 1000)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

d. $$y = x^3 - 5x$$

$y = x^3 - 5x$
$y' = (x^3 - 5x)' = 3x^2 - 5$  

*Необходимое условие экстремума.*  
$3x^2 - 5 = 0$  
$x^2 = \frac{5}{3}$  
$x_1 = -\sqrt{\frac{5}{3}} \approx -1,3$  
$x_2 = \sqrt{\frac{5}{3}} \approx 1,3$  
Необходмое условие выполняется.  

*Достаточное условие экстремума.*  
$y'(-2) = 3 \cdot (-2)^2 - 5 = 7 > 0$  
$y'(0) = 3 \cdot (0)^2 - 5 = -5 < 0$  
$y'(2) = 3 \cdot (2)^2 - 5 = 7 > 0$  
  
Следовательно, при переходе через точки $x_1 = -\sqrt{\frac{5}{3}}$ и $x_2 = \sqrt{\frac{5}{3}}$, y'(x) меняет знак. Достаточное условие выполняется. Следовательно точки $x_1, x_2$ являются точками экстремума. Точкой максимума и минимума, соответственно.  

```python
def f(x):
    return (pow(x, 3))-5*x

x = np.linspace(-3, 3, 100)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```
