### Тема “Предел функции”

**1. Предложить пример функции, не имеющей предела в нуле и в бесконечностях.**

$y = x + \frac{1}{x}$

```python
def f(x):
    return x + (1/x)

x = np.linspace(-10, 10, 100)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

**2.  Привести пример функции, не имеющей предела в точке, но определенной в ней.**

Примером такой фунции может быть функция со скачком, например функция Хевисайда, которая определена при всех значениях х, но не имеет двухстороннего предела в точне х = 0.

$$f(x) = \left\{\begin{matrix} 0, x<0;\\1, x\geq 0\end{matrix}\right.$$



```python
def f(x):
    if x < 0:
        return 0
    else:
        return 1

x = np.linspace(-10, 10, 100)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

**4. Найти предел:**

a. $$\displaystyle \lim_{x \to 0} \frac{3x^3 - 2x^2}{4x^2}$$

$\displaystyle \lim_{x \to 0} \frac{3x^3 - 2x^2}{4x^2} = \left(\frac{0}{0} \right) = \displaystyle \lim_{x \to 0} \frac{x^2*(3x - 2)}{4*x^2} = \displaystyle \lim_{x \to 0} \frac{3x - 2}{4} = -\frac{1}{2}$

b. $$\displaystyle \lim_{x \to 0} \frac{\sqrt{1+x}-1}{\sqrt[3]{1+x}-1} $$

Для нахождения предела используем формулы разности квадратов $a^2 - b^2 = (a-b)(a+b)$ и разности кубов $a^3 - b^3 = (a-b)(a^2 + ab + b^2)$. 

$\displaystyle \lim_{x \to 0} \frac{\sqrt{1+x}-1}{\sqrt[3]{1+x}-1} = \left(\frac{0}{0} \right) = \displaystyle \lim_{x \to 0} \frac{(\sqrt{1+x}-1)*(\sqrt{1+x}+1)*(\sqrt[3]{(1+x)^2} + \sqrt[3]{1+x} + 1)}{(\sqrt[3]{1+x}-1)*(\sqrt{1+x}+1)*(\sqrt[3]{(1+x)^2} + \sqrt[3]{1+x} + 1)} = \displaystyle \lim_{x \to 0} \frac{x*(\sqrt[3]{(1+x)^2} + \sqrt[3]{1+x} + 1)}{x*(\sqrt{1+x}+1)} = \displaystyle \lim_{x \to 0} \frac{(\sqrt[3]{(1+x)^2} + \sqrt[3]{1+x} + 1)}{(\sqrt{1+x}+1)} = \frac{1+1+1}{2} = \frac {3}{2}
$

### Тема “Теоремы о пределах”

**1. Найти предел:**

a. $$\displaystyle \lim_{x \to 0} \frac{sin(2x)}{4x}$$

Первый замечательный предел:  
$\displaystyle \lim_{x \to 0} \frac{sin(2x)}{4x} = \left(\frac{0}{0} \right) = \displaystyle \lim_{x \to 0} \frac{sin(2x)}{2*2x} = \frac{1}{2}*1 = \frac{1}{2} 
$

b. $$\displaystyle \lim_{x \to 0} \frac{x}{sin(x)}$$

Следствие из первого замечательного предела:  
$\displaystyle \lim_{x \to 0} \frac{x}{sin(x)} = \left(\frac{0}{0} \right) = 1$

c. $$\displaystyle \lim_{x \to 0} \frac{x}{arcsin(x)}$$

Решение произведем через замену переменных, а именно u = arcsin(x), sin(u) = x, что позволит свести выражение к первому замечательному пределу:  
$\displaystyle \lim_{x \to 0} \frac{x}{arcsin(x)} = \left(\frac{0}{0} \right) = \displaystyle \lim_{u \to 0} \frac{sin(u)}{u} = \left(\frac{0}{0} \right) = 1$
